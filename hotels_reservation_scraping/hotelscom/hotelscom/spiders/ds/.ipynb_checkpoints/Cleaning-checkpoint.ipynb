```python
import pandas as pd
```

```python
df = pd.read_json("besthoteloffersnet.json", orient='columns')
```

```python
df.head(50)
```

```python
len(df)
```

```python
df.dropna(subset=['titre', 'price','stars','id','country','city'],inplace=True)
df.drop_duplicates(subset=['id'], keep='last',inplace=True)
```

```python
len(df)
```

```python
df.info()
```

```python
def get_digit(x,y=None):
    if isinstance(x, str):
        #print(x)
        if x is not None:
            s =''
            for i in x:
                if i.isdigit():
                    s+=i
        return float(s)
    if y:
        return y
    return x
df['price'] = df['price'].apply(lambda x: get_digit(x))
df['stars'] = df['stars'].apply(lambda x: int(get_digit(x)))
df['review'] = df['review'].apply(lambda x: int(get_digit(x,'')))


```

```python
df.info()
```

```python
df.head()
```

```python
len(df)
```

```python
df['titre'] = df.titre.str.lower()
df['city'] = df.city.str.lower()
df['country'] = df.country.str.lower()
df.head()
```

```python
df.info()
```

```python
df.to_csv("hotelscom.csv")
```
